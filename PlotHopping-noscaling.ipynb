# Plot MNDO surface hopping MD results.

# First, see results averaged over all successful trajs. Note that averaging over all successful and failed trajs gave very close plots.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

name = "56tfilterb5"
timestep = 0.05  # in fs


rawe0 = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/%s/data.mean" %name, skip_header=0, skip_footer=0)
rawe1 = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/%s/data.mean1" %name, skip_header=0, skip_footer=0)
rawcoupling = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/%s/data.mean2" %name, skip_header=0, skip_footer=0)
rawpop1 = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/%s/data.mean3" %name, skip_header=0, skip_footer=0)
rawprob1 = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/%s/data.mean4" %name, skip_header=0, skip_footer=0)

time = rawe0[:,0] * timestep
E0 = rawe0[:,1]
E1 = rawe1[:,1]
coupling = rawcoupling[:,1]
pop1 = rawpop1[:,1]
prob1 = rawprob1[:,1]

plt.figure(figsize=(20,6))
plt.plot(time,E0)
plt.plot(time,E1)
plt.xlabel('time (fs)')
plt.ylabel('Potential energy (kcal/mol)')
plt.grid(True,axis='y')
plt.show()

plt.figure(figsize=(20,6))
plt.plot(time,coupling)
plt.xlabel('time (fs)')
plt.ylabel('Nonadiabatic coupling')
plt.grid(True,axis='y')
plt.show()

plt.figure(figsize=(20,6))
plt.plot(time,pop1)
plt.xlabel('time (fs)')
plt.ylabel('S0 population')
plt.grid(True,axis='y')
plt.show()

plt.figure(figsize=(20,6))
plt.plot(time,prob1)
plt.xlabel('time (fs)')
plt.ylabel('Prob. hopping to S0')
plt.grid(True,axis='y')
plt.show()
```

# Second, see single traj result, for a failed traj.:

```python
ek = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/%s/ek03.dat" %name, skip_header=1, skip_footer=0)


# ek = ek[0:80000]

DOF = 48*3-6 
temperature = ek / (0.5*DOF*0.002) # Calculate effective temperature by Ek = 1/2 NkT. R = 0.002 kcal/molK. My temperature matches MNDO temperature output.

plt.figure(figsize=(20,6))
plt.plot(ek)
plt.xlabel('time (fs)')
plt.ylabel('Kinetic Energy (kcal/mol)')
plt.grid(True,axis='y')
plt.show()

plt.figure(figsize=(20,6))
plt.plot(temperature)
plt.xlabel('time (fs)')
plt.ylabel('Effective temperature (K)')
plt.grid(True,axis='y')
plt.show()
```
