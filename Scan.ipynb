# This notebook does data analysis, following the execution of my rigidscan.sh script, which outputs several vacuum Sander energie profiles along central torsion angle of Feringa motor. Each energy profile is taken from a different puckering angle.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

filename = "/home/winter/onsager/Dih.dat"
dih = np.genfromtxt(filename, delimiter=' ', skip_header=1, skip_footer=0)

filename = "/home/winter/onsager/Etot_pucker0.dat"
Etot0 = np.genfromtxt(filename, delimiter=' ', skip_header=1, skip_footer=0)

filename = "/home/winter/onsager/Edih_pucker0.dat"
Edih0 = np.genfromtxt(filename, delimiter=' ', skip_header=1, skip_footer=0)

filename = "/home/winter/onsager/Etot_pucker30.dat"
Etot30 = np.genfromtxt(filename, delimiter=' ', skip_header=1, skip_footer=0)

filename = "/home/winter/onsager/Edih_pucker30.dat"
Edih30 = np.genfromtxt(filename, delimiter=' ', skip_header=1, skip_footer=0)


filename = "/home/winter/onsager/Etot_pucker90.dat"
Etot90 = np.genfromtxt(filename, delimiter=' ', skip_header=1, skip_footer=0)

filename = "/home/winter/onsager/Edih_pucker90.dat"
Edih90 = np.genfromtxt(filename, delimiter=' ', skip_header=1, skip_footer=0)

filename = "/home/winter/onsager/Etot_pucker180.dat"
Etot180 = np.genfromtxt(filename, delimiter=' ', skip_header=1, skip_footer=0)

filename = "/home/winter/onsager/Edih_pucker180.dat"
Edih180 = np.genfromtxt(filename, delimiter=' ', skip_header=1, skip_footer=0)


print ('Raw data length is:')
print (len(dih),len(Etot),len(Edih))
```

# Plot Edih:

```python
plt.figure(figsize=(20,10))
plt.plot(dih,Edih0)
plt.plot(dih,Edih30)
plt.plot(dih,Edih90)
plt.plot(dih,Edih180)
plt.xlim(0, 360)
plt.ylim(0, 80)
plt.xlabel('Dihedral')
plt.ylabel('Edih(kcal/mol)')
plt.legend(['pucker0', 'pucker30', 'pucker90', 'pucker180'], loc='upper left')
plt.show()
```

# Plot total energy profiles. Combined plot looked messy so I plot individually.

```python
plt.figure(figsize=(20,10))
plt.plot(dih,Etot0)
plt.xlim(0, 360)
plt.ylim(0, 1000)
plt.xlabel('Dihedral')
plt.ylabel('Etot(kcal/mol)')
plt.legend(['pucker0'], loc='upper left')
plt.show()

plt.figure(figsize=(20,10))
plt.plot(dih,Etot180)
plt.xlim(0, 360)
plt.ylim(0, 1000)
plt.xlabel('Dihedral')
plt.ylabel('Etot(kcal/mol)')
plt.legend(['pucker180'], loc='upper left')
plt.show()

plt.figure(figsize=(20,10))
plt.plot(dih,Etot90)
plt.xlim(0, 360)
plt.ylim(0, 1000)
plt.xlabel('Dihedral')
plt.ylabel('Etot(kcal/mol)')
plt.legend(['pucker90'], loc='upper left')
plt.show()

plt.figure(figsize=(20,10))
plt.plot(dih,Etot30)
plt.xlim(0, 360)
plt.ylim(0, 1000)
plt.xlabel('Dihedral')
plt.ylabel('Etot(kcal/mol)')
plt.legend(['pucker30'], loc='upper left')
plt.show()

# plt.figure(figsize=(20,10))
# plt.plot(dih,Etot0)
# plt.plot(dih,Etot30)
# plt.plot(dih,Etot90)
# plt.plot(dih,Etot180)
# plt.xlim(0, 360)
# plt.ylim(0, 1000)
# plt.xlabel('Dihedral')
# plt.ylabel('Etot(kcal/mol)')
# plt.legend(['pucker0', 'pucker30', 'pucker90', 'pucker180'], loc='upper left')
# plt.show()
```

# Basically, Etot profile of pucker(j) looks the same as profile of pucker(i), shifted in x by (j-i) degrees. Proposed explanation: (pucker=a, dih=b) is identical to (pucker=a+180, dih=b+180) by symmetry.

# Barrier along dihedral coordinate is huge, which is somewhat normal for a rigid scan. Barrier along a linear combination of pucker and dihedral coordinates could be lower, but it's hard to make a 3D energy-pucker-dihedral plot because dozens of sample points along puckering coordinate would be needed.
