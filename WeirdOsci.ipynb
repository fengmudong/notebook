# Plot MNDO surface hopping MD results.

# First, see results averaged over all successful trajs. Note that averaging over all successful and failed trajs gave very close plots.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

name = "testvel"
timestep = 0.1  # in fs

rawe0 = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/%s/data.mean" %name, skip_header=0, skip_footer=0)
rawe1 = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/%s/data.mean1" %name, skip_header=0, skip_footer=0)
rawcoupling = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/%s/data.mean2" %name, skip_header=0, skip_footer=0)
rawpop1 = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/%s/data.mean3" %name, skip_header=0, skip_footer=0)
rawprob1 = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/%s/data.mean4" %name, skip_header=0, skip_footer=0)

time = rawe0[:,0] * timestep
E0 = rawe0[:,1]
E1 = rawe1[:,1]
coupling = rawcoupling[:,1]
pop1 = rawpop1[:,1]
prob1 = rawprob1[:,1]

plt.figure(figsize=(20,6))
plt.plot(time,E0)
plt.plot(time,E1)
plt.xlabel('time (fs)')
plt.ylabel('Potential energy (kcal/mol)')
plt.grid(True,axis='y')
plt.show()

plt.figure(figsize=(20,6))
plt.plot(time,coupling)
plt.xlabel('time (fs)')
plt.ylabel('Nonadiabatic coupling')
plt.grid(True,axis='y')
plt.show()

plt.figure(figsize=(20,6))
plt.plot(time,pop1)
plt.xlabel('time (fs)')
plt.ylabel('S0 population')
plt.grid(True,axis='y')
plt.show()

plt.figure(figsize=(20,6))
plt.plot(time,prob1)
plt.xlabel('time (fs)')
plt.ylabel('Prob. hopping to S0')
plt.grid(True,axis='y')
plt.show()
```

# Second, see result of a single traj.:
## Dihedral and pyramidalization along the traj. First plot is before hopping and 2nd after hopping:

```python
import numpy as np
import matplotlib.pyplot as plt
import glob

for filename in glob.iglob("/home/winter/onsager/tsqm/dynamics/%s/run*_dihpyr.dat" %name):
    print(filename)
    hopfilename = filename[0:-11] + ".hopframe"
    hopframe = int(np.genfromtxt(hopfilename, skip_header=0, skip_footer=0, usecols=0))
    print("Hopping occurred at frame:", hopframe)
    rdih = np.genfromtxt(filename, skip_header=0, skip_footer=0, usecols=0)
    pyr = np.genfromtxt(filename, skip_header=0, skip_footer=0, usecols=1)

    # Phase unwrapping. Note that if input data is too bad, i.e. too much oscillation between like -179 and 179, phase unwrapping would fail
    dih = np.copy(rdih)
    for i in range(1,rdih.size):
        diff=dih[i]-dih[i-1]
        if diff > 180:
            dih[i:] -= 360
        elif diff < -180:
            dih[i:] += 360

    dih1=dih[0:hopframe]
    pyr1=pyr[0:hopframe]
    dih2=dih[hopframe:-1]
    pyr2=pyr[hopframe:-1]

    plt.figure(figsize=(9,5))
    plt.plot(dih1,pyr1)
    plt.plot(dih2,pyr2)
    plt.xlabel('central torsion')
    plt.ylabel('pyramidalization')
    plt.xlim(-180,0)
    plt.ylim(-50,50)
    plt.legend(['on S1', 'on S0'], loc='upper left')
    plt.show()
```
