# Comparing performance of different Nose-hoover thermostat parameters in my MNDO S0 sampling runs.

```python
import numpy as np
import matplotlib.pyplot as plt

length = 2999

nvt5 = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/nvt5.dat", skip_header=1, skip_footer=0)
nvt5 = nvt5[2000:length]
nvt1 = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/nvt1.dat", skip_header=1, skip_footer=0)
nvt1 = nvt1[2000:length]
nvt2 = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/nvt2.dat", skip_header=1, skip_footer=0)
nvt2 = nvt2[2000:length]
nvt6 = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/nvt6.dat", skip_header=1, skip_footer=0)
nvt6 = nvt6[0:length]
nvt3 = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/nvt3.dat", skip_header=1, skip_footer=0)
nvt3 = nvt3[0:length]
nvt4 = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/nvt4.dat", skip_header=1, skip_footer=0)
nvt4 = nvt4[0:length]

plt.figure(figsize=(20,6))
plt.plot(nvt5)
plt.xlabel('time / 5fs')
plt.ylabel('Temperature / K')
plt.ylim(0,500)
plt.grid(True,axis='y')
plt.legend(['tau=400fs'], loc='upper left')
plt.show()

plt.figure(figsize=(20,6))
plt.plot(nvt1)
plt.xlabel('time / 5fs')
plt.ylabel('Temperature / K')
plt.ylim(0,500)
plt.grid(True,axis='y')
plt.legend(['tau=100fs'], loc='upper left')
plt.show()

plt.figure(figsize=(20,6))

plt.plot(nvt2)
plt.xlabel('time / 5fs')
plt.ylabel('Temperature / K')
plt.ylim(0,500)
plt.grid(True,axis='y')
plt.legend(['tau=10fs'], loc='upper left')
plt.show()

plt.figure(figsize=(20,6))
plt.plot(nvt6)
plt.xlabel('time / 5fs')
plt.ylabel('Temperature / K')
plt.ylim(0,500)
plt.grid(True,axis='y')
plt.legend(['tau=400fs'], loc='upper left')
plt.show()

plt.figure(figsize=(20,6))
plt.plot(nvt3)
plt.xlabel('time / 5fs')
plt.ylabel('Temperature / K')
plt.ylim(0,500)
plt.grid(True,axis='y')
plt.legend(['tau=100fs'], loc='upper left')
plt.show()

plt.figure(figsize=(20,6))
plt.plot(nvt4)
plt.xlabel('time / 5fs')
plt.ylabel('Temperature / K')
plt.ylim(0,500)
plt.grid(True,axis='y')
plt.legend(['tau=10fs'], loc='upper left')
plt.show()
```
