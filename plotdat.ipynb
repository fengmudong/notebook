```python
import numpy as np
import matplotlib.pyplot as plt

dih = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/56t1/run0002_dih.dat", skip_header=1, skip_footer=0, usecols=1)
print(dih.shape)

pyr = np.genfromtxt("/home/winter/onsager/tsqm/dynamics/56t1/run0002_pyr.dat", skip_header=1, skip_footer=0, usecols=1)
print(pyr.shape)

# Phase unwrapping. Note that if input data is too bad, i.e. too much oscillation between like -179 and 179, phase unwrapping would fail
con = np.copy(dih)
for i in range(1,dih.size): 
    diff=con[i]-con[i-1]
    if diff > 180:
        con[i:] -= 360
    elif diff < -180:
        con[i:] += 360

plt.figure(figsize=(10,10))
plt.plot(con,pyr)
plt.xlabel('central torsion')
plt.ylabel('pyramidalization')
plt.xlim(-180,0)
plt.ylim(-90,90)
plt.grid(True,axis='y')
plt.show()
```
